## Relleno nulos de Argentina

##### Analizando las columnas donde Argentina tenia nulos. Decidi que la mejor funcion en estos casos es ffill(), ya que se tratan de columnas temporales. En estos casos utilizar el ejemplo del dia anterior es lo mas optimo.

### Creo una funcion que reciba un DataFrame y me devuelva un DataFrame con la estadistica correspondiente

```python
def estadistica_descriptiva(df):
    resultados = []
    for column in df.columns:
        if df[column].dtype in ['float64', 'int64']:  # solo tomo las columnas numericas
            resultados.append({  #genero una lista de diccionarios para transformarlo en un DataFrame mas adelante
                'Columna':column,
                'Media': df[column].mean(),
                'Mediana': df[column].median(),
                'Moda': df[column].mode()[0],
                'Varianza': df[column].var(),
                'Desviacion estandar': df[column].std(),
                'Rango': df[column].max() - df[column].min()
            })
    
    return pd.DataFrame(resultados) # Al terminar la lista, la transformo en un DataFrame

tabla_estadistica = estadistica_descriptiva(data_filtrada)  #Le aplico la funcion a nuestros datos
tabla_estadistica
```

Qu implican estas mtricas y cmo pueden ayudar en el anlisis de datos?
  Estas metricas pueden ayudar a detectar valores atipicos o inconsistencias, comparar variables entre s y elegir una estrategia de analisis o visualizacion

Se muestran todas las estadsticas en todas las columnas durante el anlisis?
  En este caso solo muestro columnas numericas. Entonces, columnas categoricas o de texto no estan dentro del analisis, debido a que metricas utilizadas no tienen sentido en ellas.

Cul es la razn de la respuesta anterior y cmo podra afectar la interpretacin de los resultados obtenidos?
  Algunas estadisticas solo tienen sentido para cierto tipos de datos, aplicar metricas numericas a columnas categoricas puede concluir en errores o interpretaciones incorrectas.

Qu representa la mediana?
    La mediana es el valor central de una columna cuando todos los datos estn ordenados de menor a mayor.

Cmo vara la dispersin de los datos en el conjunto de datos analizado, en trminos de la varianza y el rango? 
    Varianza: mide cunto se alejan los datos de la media. Valores altos indican gran dispersin. Por ejemplo, 'cumulative_confirmed' tiene una varianza altsima, indicando que los datos estn muy dispersos.
    Rango: diferencia entre el valor mximo y mnimo. Por ejemplo, 'cumulative_confirmed' tiene un rango de ~34.5 millones, lo que confirma que hay valores extremos muy altos y bajos.
    
Qu nos puede indicar esto sobre la consistencia o la variabilidad de los datos en relacin con la mediana?
    Cuando la media y la mediana estn muy separadas, suele haber valores atpicos o distribuciones sesgadas. Columnas con gran rango y varianza tienen alta variabilidad, lo que puede dificultar interpretar la tendencia central solo con la media. En cambio, columnas con varianza baja y mediana cercana a la media indican datos ms consistentes y homogneos, y la mediana representa bien el valor tpico.

# Anlisis Integrado: Tendencias, Correlaciones y Factores Contextuales

El anlisis conjunto de los datos de los pases latinoamericanos permite observar comportamientos diferenciados en la evolucin de la pandemia, as como la influencia relativa de las campaas de vacunacin, la estructura poblacional y las condiciones ambientales.

---

## Evolucin temporal y dinmica de contagios

Durante el perodo analizado (**20212022**), **Brasil** se mantuvo como el pas con mayor cantidad de casos confirmados y muertes acumuladas, marcando una diferencia significativa con el resto de la regin.  

- **Argentina** experiment su principal incremento de contagios a comienzos de 2022, con una tendencia sostenida de casos altos a lo largo del ao.  
- **Colombia**, por su parte, alcanz su pico de casos en la primera mitad de 2021, estabilizndose despus, aunque manteniendo un nivel alto de mortalidad durante todo el periodo.  
- **Chile** y **Mxico** mostraron un mejor control epidemiolgico: ambos presentaron picos ms acotados, y en el caso de Mxico incluso se registr una fuerte reduccin de casos y muertes a comienzos de 2022.  
- **Per** evidenci un crecimiento sostenido de casos que alcanz su punto mximo hacia finales de 2022, acompandose de un aumento paralelo en la mortalidad.  

---

## Vacunacin y su relacin con los indicadores sanitarios

El anlisis de correlaciones sugiere que la vacunacin tuvo un efecto mitigador sobre la mortalidad y los nuevos contagios, aunque con distinta intensidad segn el pas.  

- En **Colombia** y **Mxico**, las correlaciones negativas entre dosis acumuladas y muertes (-0.74 y -0.7 respectivamente) reflejan una posible reduccin del impacto letal del virus a medida que avanz la inmunizacin.  
- En contraste, **Argentina**, **Chile** y **Brasil** muestran altas correlaciones positivas (0.9) entre casos acumulados, muertes y dosis administradas, lo que probablemente se deba a que los aumentos de vacunacin ocurrieron en paralelo al avance de los contagios, no necesariamente antes.  

A nivel total, **Brasil** fue el pas con la mayor cantidad de dosis administradas, mientras que **Chile** tuvo la menor, con una diferencia aproximada de **285 millones de dosis**.  
Esta brecha refleja tanto la magnitud poblacional como la capacidad logstica de cada pas.

---

## Factores ambientales

El estudio de las variables climticas no mostr correlaciones significativas entre la **temperatura media** y las cifras de **casos** o **muertes**.  
Esto sugiere que, en el contexto latinoamericano, la temperatura no fue un factor determinante en la propagacin o gravedad de la pandemia.  

Sin embargo, no puede descartarse su influencia indirecta, asociada a **comportamientos estacionales** o **patrones de movilidad**.

---

## Estructura poblacional y vulnerabilidad

El anlisis demogrfico revela diferencias estructurales que pueden influir en la vulnerabilidad sanitaria.  

- **Argentina** y **Per** poseen una poblacin predominantemente joven, con ms del **50 % menor de 30 aos** y menos del **15 % mayor de 60**, lo cual puede haber contribuido a tasas de mortalidad moderadas.  
- **Mxico** tambin presenta una pirmide poblacional joven, con un **40 % de personas menores de 20 aos**.  
- En cambio, **Chile** muestra una distribucin ms equilibrada entre los distintos grupos etarios, mientras que **Brasil** y **Colombia** concentran la mayora de su poblacin en edades productivas (**2050 aos**).

---

## Tasa de mortalidad y diferencias por gnero

La **tasa de mortalidad masculina** resulta aproximadamente un **85 % ms alta** que la femenina, lo que confirma una tendencia observada tambin a nivel global.  

Esto podra estar relacionado con factores **biolgicos** (como respuesta inmune) y **socioculturales** (mayor exposicin laboral o comorbilidades).
